```python
import pandas as pd
import numpy as np
import joblib
from sklearn.linear_model import PoissonRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# 1. Cargar los datos ya preparados (el CSV con elo y pesos)
datos = pd.read_csv("data/DataF.csv")
datos["date"] = pd.to_datetime(datos["date"])

# --- 2. Definir X, y, pesos ---
features = ["mi_elo", "elo_rival", "es_local"]
X = datos[features]
y = datos["goles_metidos"]
pesos = datos["peso"]

# --- 3. Crear el pipeline con los MEJORES hiperparámetros encontrados ---
modelo = Pipeline([
    ("scaler", StandardScaler()),
    ("poisson", PoissonRegressor(alpha=0.0)),   # alpha=0.0 ganó en el GridSearch
])

# --- 4. Entrenar con TODOS los datos (ya no separamos train/test:
#        eso era para comparar modelos; ahora que elegimos, usamos todo) ---
modelo.fit(X, y, poisson__sample_weight=pesos)

# --- 5. Guardar el modelo entrenado a disco ---
joblib.dump(modelo, "modelo_poisson.joblib")
print("Modelo guardado en modelo_poisson.joblib")

# --- 6. Guardar también el diccionario de Elo (lo necesita la predicción) ---
#        Lo reconstruimos desde los datos, o si lo tienes en un dict, guárdalo.
#        Aquí asumo que tienes elo_fifa en este script; si no, cópialo.
joblib.dump(elo_fifa, "elo_fifa.joblib")
print("Elo guardado en elo_fifa.joblib")
#Tambien saco el elo porque cuando el usuario escriba el nombre de un equipo, lo busco en el diccionario aqui.
```
